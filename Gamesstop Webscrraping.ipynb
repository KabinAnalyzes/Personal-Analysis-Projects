```python
import os
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
baseurl = 'https://www.gamestop.com/'
accessurl = 'https://www.gamestop.com/video-games/playstation-5/games'
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'
}
listo = [0, 24, 48, 72, 96, 120, 144]
productlink= []
for x in listo:
    r = requests.get(f'https://www.gamestop.com/video-games/playstation-5/games?start={x}&sz=24', headers=headers)
    soup = bs(r.content, 'lxml')
    productlist= soup.find_all('div', class_='product-tile product-detail')
    for item in productlist:
        for link in item.find_all('a', href=True):
            productlink.append(baseurl + link['href'])
gameslist=[]         
for links in productlink:
    r = requests.get(links, headers=headers)
    soup = bs(r.content, 'lxml')
    name =soup.find('div', class_='product-name-section').get_text(strip= True)
    price = soup.find_all('span', class_='actual-price')[-1].get_text(strip= True)
    try:
        publisher = soup.find('span', class_='product-publisher').get_text(strip=True)
    except:
        publisher = 'NA'
    try:
        rating = soup.find('h1', class_='rating-value').get_text(strip=True)
    except:
        rating= 'no rating'
    try:
        reviews = soup.find('div', class_='review-count').get_text(strip=True)
    except:
        reviews = 'no reviews'

    games = {
        'name': name,
        'publisher': publisher,  
        'price':price,
        'rating':rating,
        'reviewer_count': reviews    
    }
    games
    gameslist.append(games)
    print('Saving: ', games['name'])

df = pd.DataFrame(gameslist)
print(df.head(5))
```
