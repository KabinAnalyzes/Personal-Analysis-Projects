```python
baseurl= 'https://www.imdb.com/'
accessurl ='https://www.imdb.com/chart/top/'
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'
}
movielink =[]
r = requests.get(accessurl, headers=headers)
soup = bs(r.content , 'lxml')
movielist = soup.find_all('td' , class_='titleColumn')
for item in storelist:
        for link in item.find_all('a', href=True):
            movielink.append(baseurl + link['href'])
Mlist= []
for links in movielink:
    r = requests.get(links, headers=headers)
    soup = bs(r.content, 'lxml')
    name =soup.find('h1', class_='TitleHeader__TitleText-sc-1wu6n3d-0').get_text(strip=True)
    IMDB_Rating = soup.find('span', class_='AggregateRatingButton__RatingScore-sc-1ll29m0-1').get_text(strip=True)
    try:
        Popularity = soup.find('div', class_='TrendingButton__TrendingScore-bb3vt8-1').get_text(strip=True)
    except:
        Popularity = 'NA'
    Director = soup.find('a', class_='ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link').get_text(strip=True)
    try:
        Meta_score = soup.find('span', class_='score-meta').get_text(strip=True)
    except:
        Meta_score = 'NA'
    CriticR = soup.find_all('span', class_='score')[1].get_text(strip=True)
    UserR = soup.find('span', class_='score').get_text(strip=True)
    Movies = {
        'name': name,
        'director' : Director,
        'Stars': IMDB_Rating, 
        'popularity' : Popularity,
        'Metascore' : Meta_score, 
        'User_Reviews': UserR,
        'Critic_Reviews': CriticR,
    }
    Mlist.append(Movies)
    print('Mining: ', Movies['name'])
df = pd.DataFrame(Mlist)
print(df.head(5))
```
